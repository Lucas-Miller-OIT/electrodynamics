```python
## TRY THIS CODE BUT WITH EXACT LOCATIONS FOR EACH MAGNET (S1, S2.. S16)
## USE s4 = Box(mag=M, dim=D, pos = [ 4, 0, -4], angle=45, axis=[1,0,0]) TO SET DIFFERENT ANGLES
## Documentation: https://magpylib.readthedocs.io/en/latest/_autogen/magpylib.core.html

# imports
import numpy as np
import matplotlib.pyplot as plt
import magpylib as magpy

## Use this command to turn ON interactive plots:
%matplotlib notebook

## Use this command to turn OFF interactive plots.
# %matplotlib inline

# create figure
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121, projection='3d') # 3D plotting axis
ax2 = fig.add_subplot(122) # 2D plotting axis

# create magnets
# mag = magnetization vector magnitude + direction (units = mT)
# dim = shape dimensions
# pos = location of the center of the shape
# angle = angle relative to the x-axis

s1 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1,1.5,2], angle=-90)
s2 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-3,1.5,2], angle=90)
s3 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1.5,3,2], angle=-180)
s4 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1.5,1,2])
s5 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1,1.5,2], angle=-90)
s6 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[3,1.5,2], angle=90)
s7 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1.5,1,2])
s8 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1.5,3,2], angle=-180)
s9 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1,-1.5,2], angle=90)
s10 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[3,-1.5,2], angle=90)
s11 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1.5,-1,2])
s12 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[1.5,-3,2])
s13 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1,-1.5,2], angle=-90)
s14 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-3,-1.5,2], angle=90)
s15 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1.5,-1,2])
s16 = magpy.source.magnet.Box(mag=[10,0,0],dim=[0.4,0.2,3],pos=[-1.5,-3,2])

# create collection
c = magpy.Collection(s1,s2, s3, s4, s5, s6, s7, s8, s9, s10, s11, s12, s13, s14, s15, s16)

# display system geometry on ax1
magpy.displaySystem(c,subplotAx=ax1,suppress=True)

# calculate B-field on a grid
xs = np.linspace(-10,10,29)
zs = np.linspace(-10,10,29)
Bs = np.array([[c.getB([x,0,z]) for x in xs] for z in zs])

# display field in xz-plane using matplotlib
X,Z = np.meshgrid(xs,zs)
U,V = Bs[:,:,0], Bs[:,:,2]
ax2.streamplot(X, Z, U, V, color=np.log(U**2+V**2),density=2)
plt.show()

# display collection
magpy.displaySystem(c,direc=True,figsize=(6,6))

# use U, V and weeks 1,2 to plot field lines in 3D
# Week 1 3D
# U,V,W = Bs[:,:,0], Bs[:,:,2], Bs[:,:,1]
```

```python
from magpylib.source.magnet import Box
import magpylib as magpy
# fixed magnet parameters
M = [1,0,0] #magnetization
D = [1,2,4] #dimension
# magnets with Euler angle orientations
s1 = Box(mag=M, dim=D, pos = [-4, 0, 4])
s2 = Box(mag=M, dim=D, pos = [ 4, 0, 4])
s3 = Box(mag=M, dim=D, pos = [-4, 0, -4])
s4 = Box(mag=M, dim=D, pos = [ 4, 0, -4], angle=45, axis=[1,0,0])
# collection
c = magpy.Collection(s1,s2,s3,s4)
# display collection
magpy.displaySystem(c,direc=True,figsize=(6,6))
```
